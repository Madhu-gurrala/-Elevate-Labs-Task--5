## Removing the unwanted columns for data Analysis
## Syntax:Dataframe_name.drop(['col1_name',....],axis=1,inplace=True)

```python
#  Removing the unwanted columns for data analysis
sales.drop(['year','month','full_name','User Name'],axis=1,inplace=True)
```

```python
# Creating a new column from existing columns
## Combining all into the single column
sales['customer_name']=pd.concat([sales['Name Prefix'],sales['Middle Initial'],sales['First Name'],sales['Last Name']],axis=1).agg(' '.join,axis=1)
## Dropping the original columns
sales.drop(['Name Prefix','Middle Initial','First Name','Last Name'],axis=1,inplace=True)
## Print the data frame
sales
```
